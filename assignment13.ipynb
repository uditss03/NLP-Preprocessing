```python
import re
import string
import nltk
from nltk.tokenize import word_tokenize
#nltk.download('punkt')

# URLS
def remove_urls (vTEXT):
    vTEXT = re.sub(r'(https|http)?:\/\/(\w|\.|\/|\?|\=|\&|\%)*\b', '', vTEXT, flags=re.MULTILINE)
    return(vTEXT)
print(remove_urls("hello .world https://github.com"))


# digits
def remove_numbers(text):
    text = re.sub(r'\d+', ' ', text)
    return text
print(remove_numbers("hello . worlds 2 dsms 3 s42"))

# remove ' [ ] - -- etc
def remove_replace(text):
    text = text.replace('-',' ')
    text = text.replace('--',' ')
    text = text.replace('[',' ')
    text = text.replace(']',' ')
    text = text.replace("'",' ')
    return text
print(remove_replace("hellwo . dnfj --df-fd ' dds [sdd]  rere "))

# remove non-aplhabets
def remove_nonalphabets(text):
    text = re.sub("[^a-zA-Z]+", " ", text)
    return text
print(remove_nonalphabets("hellow . * 23 g mfng }"))

# remove stopword

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    result = [i for i in tokens if not i in stop_words]
    return result
print(remove_stopwords("India is the best Country"))
```

```python
corpus = []
for review in reviews:
    review = remove_nonalphabets(review)
    review = remove_urls(review)
    review = remove_numbers(review)
    review = remove_replace(review)
    review = remove_stopwords(review)
    corpus.append(review)
```
